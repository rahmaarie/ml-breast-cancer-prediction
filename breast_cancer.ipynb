# **1. Install Library**

```python
# Library ini diperlukan untuk memanipulasi data, membangun model
import pandas as pd
```

# **2. Read Dataset**

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer() #Memuat dataset dari Scikit-Learn
x = cancer.data  # Fitur
y = cancer.target  # Label

df_x = pd.DataFrame(x, columns=cancer.feature_names)
df_y = pd.Series(y, name='target')

```

```python
# Menampilkan data fitur dalam bentuk DataFrame
df_x
```

```python
# Menampilkan data fitur dalam bentuk Series
df_y
```

```python
# Gabungkan fitur dan target dalam satu Dataframe
df = pd.concat([df_x, df_y], axis=1)

# Menampilkan 25 baris dari dataset
df.head(25)
```

```python
df.info()
```

```python
# Menampilkan nilai unik dari target
df['target'].unique()
```

```python
df.describe()
```

# **3. Split Data**

```python
# Membagi dataset menjadi data latih dan data uji(test_size) dengan proporsi 70:30.
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

# **4. Train the Model**

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Membangun dan melatih model Random Forest dengan 100 pohon keputusan.
model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(x_train, y_train)

# Memprediksi label pada data uji dan mengevaluasi performa model menggunakan akurasi.
y_pred_rf = model_rf.predict(x_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Akurasi: {accuracy_rf * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_rf)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_rf,target_names=cancer.target_names))
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

logistic_model = LogisticRegression(max_iter=3000) #bisa menaikan jumlah iterasi jika nilai iterasinya belum sesuai
logistic_model.fit(x_train,y_train)


# Memprediksi label pada data uji dan mengevaluasi performa model menggunakan akurasi.
y_pred_logistic = logistic_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred_logistic)

print("Akurasi:",accuracy)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_logistic)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_logistic,target_names=cancer.target_names))
```

**Support Vector Machine**

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm_model = SVC(kernel='linear')
svm_model.fit(x_train, y_train)


# Memprediksi label pada data uji dan mengevaluasi performa model menggunakan akurasi.
y_pred_SVM = svm_model.predict(x_test)

accuracy_svm = accuracy_score(y_test, y_pred_SVM)

print(f"Akurasi: {accuracy_svm:.4f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_SVM)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_SVM,target_names=cancer.target_names))
```

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

nb_model = GaussianNB()
nb_model.fit(x_train, y_train)


# Memprediksi label pada data uji dan mengevaluasi performa model menggunakan akurasi.
y_pred_NB = nb_model.predict(x_test)

accuracy_nb = accuracy_score(y_test, y_pred_NB)

print("Akurasi :",accuracy_nb)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_NB)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_NB,target_names=cancer.target_names))
```

**K-Nearest Neightbor(KNN)**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train,y_train)

# Memprediksi label pada data uji dan mengevaluasi performa model menggunakan akurasi.
y_pred_knn = knn_model.predict(x_test)

accuracy_knn = accuracy_score(y_test, y_pred_knn)

print("Akurasi:",accuracy_knn)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_knn)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# **5. Predict & Evaluate**

```python

models = ['Random Forest', 'SVM', 'Naive Bayes', 'KNN']
accuracies = [accuracy_rf, accuracy_svm, accuracy_nb, accuracy_knn]


plt.figure(figsize=(8, 9))
bars = plt.bar(models, accuracies, color='lightblue')

plt.title('Akurasi Model Klasifikasi pada Dataset Breast Cancer', fontsize=14)
plt.xlabel('Model', fontsize=16)
plt.ylabel('Akurasi', fontsize=12)
plt.ylim(0, 1)


for bar, accuracy in zip(bars, accuracies):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.01,
             f'{accuracy:.2f}', ha='center', va='bottom', color='black', fontsize=10)

plt.show()
```
